## Data Wrangling

> After observing the dataset and proposed questions for the analysis we will be keeping only relevent data deleting the unsued data so that we can make our calculation easy and understandable. 
> .



```python
# Load your data

data = pd.read_csv('tmdb-movies.csv')
data.head(2)
```

```python
print(data.shape)
```

```python
print(list(data.columns.values))
```

```python
genres = []
for item in data:
    genres.append(data.genres.str.split('|').tolist())
```

```python
genres =['0' if x is np.nan else x for x in genres[0]]
```

```python
len(genres)
```

```python
genres_data = pd.DataFrame.from_records(genres, columns = ['genres_1', 'genres_2','genres_3','genres_4','genres_5'])
genres_data
```

```python
data = pd.concat([data, genres_data], axis=1)
data
```

```python
cast = []
for item in data:
    cast.append(data.cast.str.split('|').tolist())
cast[0][0]
```

```python
cast =['0' if x is np.nan else x for x in cast[0]]
```

```python
cast_data = pd.DataFrame.from_records(cast, columns = ['cast_1', 'cast_2','cast_3','cast_4','cast_5'])
cast_data
```

```python
data = pd.concat([data, cast_data], axis=1)
data
```

```python
keywords = []
for item in data:
    keywords.append(data.keywords.str.split('|').tolist())
keywords[0][0]
```

```python
keywords =['0' if x is np.nan else x for x in keywords[0]]
```

```python
keywords_data = pd.DataFrame.from_records(keywords, columns = ['keyword_1', 'keyword_2','keyword_3','keyword_4','keyword_5'])
keywords_data
```

```python
data = pd.concat([data, keywords_data], axis=1)
data
```

```python
data.release_date
```

```python
date = []
for item in data:
    date.append(data.release_date.str.split('/').tolist())
date[0][0]
```

```python
date_data = pd.DataFrame.from_records(date[0], columns = ['release_month', 'release_day','release'])
date_data
```

```python
data = pd.concat([data, date_data], axis=1)
data
```

```python
data.drop(['cast','keywords','genres','imdb_id','homepage','release_date','release'],axis =1,inplace = True)
```

```python
#number of unique values in each column
data.nunique()
```

```python
data.info()
```

```python
print('number of duplicated rows :',sum(data.duplicated()))
data.drop_duplicates(inplace=True)
print('number of duplicated rows after drop_duplicates:',sum(data.duplicated()))
```

```python
data.describe()
```

```python
new_labels = []
for col in data.columns:
    if '_adj' in col:
        new_labels.append(col[:-4])  # exclude last 3 characters
    else:
        new_labels.append(col)

# new labels for our columns
new_labels
```

```python
data.columns = new_labels
data.columns
```

```python
data['release_month']= data['release_month'].astype(int)
```
